```
################################################################################
# FEDERAL UNIVERSITY OF CATALÃO (UFCAT)
# DEVELOPERS:
# DR. WANDERLEI MALAQUIAS PEREIRA JUNIOR (UFCAT / FENG)
# DR. FRAN SÉRGIO LOBATO (UFU / FEQUI)
# GITHUB: wwww.linklink.com.br
# LICENSE: aqui verificar
################################################################################

################################################################################
# DESCRIPTION SCRIPT:

# LIBRARY OF GENDER FIREFLY ALGORITHM OPTIMIZATION ALGORITHM (GDFA)
################################################################################

################################################################################
# LIBRARIES
# NATIVE PYTHON
import numpy as np
import random as rnd

# GPEE DEVELOPERS
# EVALUATION FITNESS
def FIT_VALUE(OF_I):
    if OF_I >= 0:
        FIT_I = 1 / (1 + OF_I)
    elif OF_I < 0:
        FIT_I = 1 + abs(OF_I)
    return FIT_I

# EVALUATION OBJECTIVE FUNCTION
def OF_VALUE(X):
    X_0 = X[0]
    X_1 = X[1]
    OF_I = X_0 * X_0 + X_1 * X_1
    return OF_I

# BOOT POPULATION
def INITIAL_POPULATION(NPOP, D, X_MIN, X_MAX, TYPE_BOOT, TYPE_SEED):
    # INITIALIZATION DESIGN VARIABLES MATRIX
    X = np.zeros((NPOP, D))
    np.random.RandomState(seed = TYPE_SEED)
    # START POPULATION
    if TYPE_BOOT == 'UNIFORM':
        for I_COUNT in range(N_POP):
            for J_COUNT in range(D):
                RANDON_NUMBER = np.random.random()
                X[I_COUNT, J_COUNT] = X_MIN[J_COUNT] + (X_MAX[J_COUNT] - X_MIN[J_COUNT]) * RANDON_NUMBER
    return X

# CALCULATION OF THE DISCRIMINATING FACTOR OF THE MALE AND FEMALE FIREFLIES POPULATION
def DISCRIMINANT_FACTOR_MALE_MOVIMENT(FIT_I_MALE_FIREFLY_I, FIT_I_FEMALE_FIREFLY_J):
    # COMPARISON OF FIREFLY BRIGHTNESS
    if FIT_I_MALE_FIREFLY_I > FIT_I_FEMALE_FIREFLY_J:
        FACTOR_D = 1
    else:
        FACTOR_D = -1
    return FACTOR_D

# CALCULATION OF ATTRACTIVENESS AMONG FIREFLIES
def ATTRACTIVENESS_FIREFLY_PARAMETER(BETA_0, GAMMA, X_I, X_J, D):
    # DISTANCE BETWEEN X_I AND X_J FIREFLIES
    VALUE_1 = 0
    for I_COUNT in range(D):
        VALUE_1 += (X_I[I_COUNT] - X_J[I_COUNT]) ** 2
    R_IJ = np.sqrt(VALUE_1)
    # BETA ATTRACTIVENESS
    BETA_0 = []
    for J_COUNT in range(D):
        BETA = np.append(BETA_0 * EXP(- GAMMA[J_COUNT] * R_IJ))
    return BETA

# MALE FIREFLY MOVEMENT
def MALE_FIREFLY_MOVEMENT(X_OLD_I_MALE, FIT_I_MALE, NPOP_FEMALE, Y_FEMALE, FIT_FEMALE, BETA_0, GAMMA, D):
    # INITIALIZATION VARIABLES
    SECOND_TERM = []
    THIRD_TERM = []
    X_NEW_I_MALE  []
    # FIRST RANDOM FEMALE FIRELY SELECT
    FIRST_RANDOM_FEMALE_FIREFLY = rnd.randint(0, (NPOP_FEMALE - 1))
    Y_K_FEMALE = Y_FEMALE[FIRST_RANDOM_FEMALE_FIREFLY, :]
    FIT_I_FEMALE_FIREFLY_K = FIT_FEMALE[FIRST_RANDOM_FEMALE_FIREFLY]
    # DISCRIMINANT D FACTOR
    FACTOR_D_1 = DISCRIMINANT_FACTOR_MALE_MOVIMENT(FIT_I_MALE_FIREFLY, FIT_I_FEMALE_FIREFLY_K)
    # ATTRACTIVENESS AMONG FIREFLIES
    BETA_1 = ATTRACTIVENESS_FIREFLY_PARAMETER(BETA_0, GAMMA, X_OLD_I_MALE, Y_K_FEMALE, D)
    # LAMBDA RANDOM PARAMETER
    LAMBDA = np.random.random()
    for I_COUNT in range(D):
        SECOND_TERM = np.append(FACTOR_D_1 * BETA_1[I_COUNT] * LAMBDA * (Y_K_FEMALE[I_COUNT] - X_OLD_I_MALE[I_COUNT])
    # SECOND RANDOM FEMALE FIRELY SELECT
    SECOND_RANDOM_FEMALE_FIREFLY = rnd.randint(0, (NPOP_FEMALE - 1)) # aqui eu tenho que fazer uma forma de não pegar a mesma fêmea portanto first and second devem ser diferentes
    Y_J_FEMALE = Y_FEMALE[SECOND_RANDOM_FEMALE_FIREFLY, :]
    FIT_I_FEMALE_FIREFLY_J = FIT_FEMALE[SECOND_RANDOM_FEMALE_FIREFLY]
    # DISCRIMINANT D FACTOR
    FACTOR_D_2 = DISCRIMINANT_FACTOR_MALE_MOVIMENT(FIT_I_MALE, FIT_I_FEMALE_FIREFLY_J)
    # ATTRACTIVENESS AMONG FIREFLIES
    BETA_2 = ATTRACTIVENESS_FIREFLY_PARAMETER(BETA_0, GAMMA, X_I_MALE, Y_J_FEMALE, D)
    # MU RANDOM PARAMETER
    MU = np.random.random()
    for J_COUNT in range(D):
        THIRD_TERM = np.append(FACTOR_D_2 * BETA_2[J_COUNT] * MU * (Y_J_FEMALE[J_COUNT] - X_OLD_I_MALE[J_COUNT])
    # UPDATE MALE POSITION
    for K_COUNT in range(D):
        X_NEW_I_MALE[K_COUNT] = X_OLD_I_MALE[K_COUNT] + SECOND_TERM[K_COUNT] + THIRD_TERM[K_COUNT]
    return X_NEW_I_MALE

# FEMALE FIREFLY MOVEMENT
def FEMALE_FIREFLY_MOVEMENT(Y_OLD_I_FEMALE, X_BEST, FIT_BEST, BETA_0, GAMMA, D):
    # INITIALIZATION VARIABLES
    Y_NEW_I_FEMALE = []
    # ATTRACTIVENESS AMONG FIREFLIES (Y_FEMALE AND X_BEST)
    BETA = ATTRACTIVENESS_FIREFLY_PARAMETER(BETA_0, GAMMA, Y_OLD_I_FEMALE, X_BEST, D)
    # PHI RANDOM PARAMETER
    PHI = np.random.random()
    # UPDATE FEMALE POSITION
    for I_COUNT in range(D):
        Y_NEW_I_FEMALE = np.append(Y_OLD_I_FEMALE[I_COUNT] + BETA[I_COUNT] * PHI * (X_BEST[I_COUNT] - Y_OLD_I_FEMALE[I_COUNT])
    return Y_NEW_I_FEMALE

# BEST POSITION, OF, FIT VALUES
def BEST_VALUES(X, OF, FIT, D):
    # BEST POSITION IN POPULATION
    SORT_POSITIONS = np.argsort(OF)
    BEST_POSITION = SORT_POSITIONS[0,0]
    # G_BEST DEFINITION
    X_G_BEST = X[BEST_POSITION, :]
    OF_G_BEST = OF[BEST_POSITION, :]
    FIT_G_BEST = FIT[BEST_POSITION, :]
    return X_G_BEST, OF_G_BEST, FIT_G_BEST
```
